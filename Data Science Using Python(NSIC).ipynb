```python
## Pandas – Data Handling & Cleaning
```

```python
import pandas as pd

# Creating a DataFrame
data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [21, 22, None, 23],
    "Marks": [85, 90, 88, None]
}

df = pd.DataFrame(data)
print(df)

# Handling missing values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Marks"].fillna(df["Marks"].mean(), inplace=True)

print("\nAfter Cleaning:\n", df)
```

```python
## Exploratory Data Analysis (EDA)
```

```python
# Statistical summary
print(df.describe())

# Filtering data
high_scorers = df[df["Marks"] > 88]
print("\nHigh Scorers:\n", high_scorers)
```

```python
## Data Visualization (Matplotlib & Seaborn)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot
plt.bar(df["Name"], df["Marks"])
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.show()

# Seaborn visualization
sns.histplot(df["Marks"], kde=True)
plt.title("Marks Distribution")
plt.show()
```

```python
## Machine Learning – Clustering 
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Sample dataset
X = [[200, 20], [300, 25], [400, 30], [150, 18], [500, 35]]

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)

# Results
print("Cluster Labels:", kmeans.labels_)
print("Centroids:\n", kmeans.cluster_centers_)
```

```python
## Clustering Insight Extraction
```

```python
import numpy as np

clusters = np.array(kmeans.labels_)
for i in range(2):
    print(f"Cluster {i} contains {np.sum(clusters == i)} data points")
```

```python
## Simple Predictive Model 
```

```python
from sklearn.linear_model import LinearRegression

# Data
X = [[1], [2], [3], [4]]
y = [50, 60, 70, 80]

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction
prediction = model.predict([[5]])
print("Predicted value:", prediction[0])
```
